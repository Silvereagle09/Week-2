```python
import torch
from dataclasses import dataclass
from typing import Any, Dict, List, Union

@dataclass
class WhisperCollator:
    processor: Any

    def __call__(self, features: List[Dict[str, Union[List[int], torch.Tensor]]]) -> Dict[str, torch.Tensor]:
        input_features = torch.stack([torch.tensor(f["input_features"]) for f in features])
        labels = [torch.tensor(f["labels"]) for f in features]

        # pad labels
        labels_batch = self.processor.tokenizer.pad(
            {"input_ids": labels}, padding=True, return_tensors="pt"
        )

        labels = labels_batch["input_ids"].masked_fill(labels_batch["input_ids"] == self.processor.tokenizer.pad_token_id, -100)

        batch = {
            "input_features": input_features,
            "labels": labels,
        }
        return batch

data_collator = WhisperCollator(processor)
```

```python
import os
import pandas as pd

root_dir = "/content/Week-1/English-audio-dataset"
csv_path = "/content/Week-1/english_dataset_prepared.csv"

df = pd.read_csv(csv_path)
print(f"ğŸ“¦ Total rows in CSV: {len(df)}")

# 1ï¸âƒ£ Check for missing files
missing_files = [p for p in df["audio_path"] if not os.path.exists(p)]
if missing_files:
    print(f"âš ï¸ Missing audio files: {len(missing_files)}")
    print("Example:", missing_files[:3])
else:
    print("âœ… All audio files exist!")

# 2ï¸âƒ£ Check for duplicates
duplicates = df["audio_path"].duplicated().sum()
if duplicates:
    print(f"âš ï¸ Found {duplicates} duplicate entries in CSV.")
else:
    print("âœ… No duplicate audio entries.")

# 3ï¸âƒ£ Check for unused transcript lines
used_ids = set(os.path.basename(p).replace(".flac", "") for p in df["audio_path"])

all_ids = set()
for subdir, dirs, files in os.walk(root_dir):
    for file in files:
        if file.endswith(".trans.txt"):
            with open(os.path.join(subdir, file), "r", encoding="utf-8") as f:
                for line in f:
                    parts = line.strip().split(" ", 1)
                    if len(parts) == 2:
                        all_ids.add(parts[0])

unused_ids = all_ids - used_ids
if unused_ids:
    print(f"âš ï¸ Unused transcript lines found: {len(unused_ids)}")
    print("Example:", list(unused_ids)[:5])
else:
    print("âœ… All transcript lines used!")

# 4ï¸âƒ£ Check for missing transcripts
missing_text = df["text"].isna().sum()
if missing_text:
    print(f"âš ï¸ Missing text entries: {missing_text}")
else:
    print("âœ… All text entries present!")
```

```python
audio_path = "/content/Week-1/English-audio-dataset/1272/128104/1272-128104-0000.flac"

speech_array, sampling_rate = torchaudio.load(audio_path)
speech = torchaudio.functional.resample(speech_array, sampling_rate, 16000)

input_features = processor(
    speech.squeeze().numpy(),
    sampling_rate=16000,
    return_tensors="pt"
).input_features

with torch.no_grad():
    predicted_ids = model.generate(input_features)

transcription = processor.batch_decode(predicted_ids, skip_special_tokens=True)[0]
print("ğŸ—£ï¸ Model Transcription:\n", transcription)
```
